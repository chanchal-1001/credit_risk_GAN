```python
context = ts.context(ts.SCHEME_TYPE.CKKS, poly_modulus_degree=8192, coeff_mod_bit_sizes=[60,40,40,60])
context.generate_galois_keys() #Optional
context.generate_relin_keys() #Optional
context.global_scale = pow(2,10)
```

```python
data = df.copy()
```

```python
data = torch.tensor(data.values).float()  
```

```python
# Define the split ratio (e.g., 80% for training, 20% for testing)
split_ratio = 0.8

# Calculate the number of elements for training and testing
total_elements = data.size(0)
num_training_elements = int(total_elements * split_ratio)
num_testing_elements = total_elements - num_training_elements

# Split the tensor into training and testing subsets
training_data = data[:num_training_elements]
testing_data = data[num_training_elements:]
```

```python
X_train = training_data[:, :-1]  # All rows, all columns except the last one
Y_train = training_data[:, -1]
```

```python
encrypted_data = []
encrypted_data = [ts.ckks_vector(context, x.tolist()) for x in X_train]
```

```python
print(len(encrypted_data))
```

```python
#batch the data
batch_size = 50
num_batches = len(encrypted_data) // batch_size
encrypted_batches = [encrypted_data[i:i + batch_size] for i in range(0,len(encrypted_data), batch_size)]
```

```python
len(encrypted_batches[0])
```

```python
X_train.shape
```

```python
#Define Generator
def create_generator():
    generator = Sequential()
    generator.add(Dense(units=256, input_dim=X_train.shape[1]))
    generator.add(LeakyReLU(0.2))
    
    generator.add(Dense(units=512))
    generator.add(LeakyReLU(0.2))
    
    generator.add(Dense(units=1024))
    generator.add(LeakyReLU(0.2))
    
    generator.add(Dense(units=X_train.shape[1], activation='sigmoid'))
    
    generator.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0002, beta_1=0.5))
    return generator
```

```python
#Define discriminator
def create_discriminator():
    discriminator = Sequential()
    discriminator.add(Dense(units=256, input_dim=X_train.shape[1]))
    discriminator.add(LeakyReLU(0.2))
    discriminator.add(Dropout(0.3))
    
    discriminator.add(Dense(units=512))
    discriminator.add(LeakyReLU(0.2))
    discriminator.add(Dropout(0.3))
    
    discriminator.add(Dense(units=256))
    discriminator.add(Dense(units=0))
    discriminator.add(LeakyReLU(0.2))
    
    discriminator.add(Dense(units=X_train.shape[1], activation='sigmoid'))
    
    discriminator.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0002, beta_1=0.5))
    return discriminator
```

```python
# create GAN
def create_gan(discriminator, generator):
    discriminator.trainable = False
    gan_input = Input(shape=(X_train.shape[1],))
    x = generator(gan_input)
    gan_output = discriminator(x)
    gan = Model(inputs=gan_input, outputs=gan_output)
    gan.compile(loss= 'binary_crossentropy', optimizer='adam')
    gan.summary()
    return gan
```

```python
#define GAN model
def train_gan(gan, generator, discriminator, epochs=1, batch_size=50):
    
    for e in range(epochs):
        for batch in encrypted_batches:
            noise = np.random.normal(0,1,[batch_size,20])
            generated_data = generator.predict(noise)
            
            #Decrypted Data
            decrypted_batch = [enc_val.decrypt() for enc_val in batch]
            real_data = Variable(torch.Tensor(decrypted_batch))
            
            
            #Encrypted Data
            #for enc_real_data in batch:
            #enc_real_data = batch
            discriminator.trainable = True
            discriminator.train_on_batch(real_data,np.ones((batch_size,20)))
            discriminator.train_on_batch(generated_data,np.zeros((batch_size,20)))
            discriminator.trainable = False
            loss = gan.train_on_batch(noise, np.ones((batch_size,20)))
            print('Loss Value: ', loss)
```

```python
#create the models
generator = create_generator()
discriminator = create_discriminator()
```

```python
#train GAN model
gan = create_gan(discriminator, generator)
train_gan(gan,generator, discriminator)
```
